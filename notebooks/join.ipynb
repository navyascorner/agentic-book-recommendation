```python
import pandas as pd
from pathlib import Path
```

```python
data_dir = Path("../data/intermediate/books-cleaned")

file_names = [
    "children.csv",
    "comics.csv",
    "fantasy.csv",
    "history.csv",
    "mystery.csv",
    "poetry.csv",
    "romance.csv",
    "young.csv",
]
```

```python
dfs = []

for file_name in file_names:
    file_path = data_dir / file_name
    
    temp_df = pd.read_csv(file_path)
    temp_df["source_file"] = file_name.replace(".csv", "")
    
    print(f"{file_name}: {temp_df.shape}")
    
    dfs.append(temp_df)

df = pd.concat(dfs, ignore_index=True)
```
